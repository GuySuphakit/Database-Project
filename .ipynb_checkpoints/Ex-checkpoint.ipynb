## 1 สร้างคำสั่งสืบค้นข้อมูลเพื่อหาจำนวนครั้งที่ลูกค้าใช้จ่ายเกิน 1,500 บาทโดยแยกตามเดือนและปี 

```python
%%sql
select extract(year from intime)::int "year",
    extract(month from intime)::int "month",
    sum(netpay)::int netpay,
    count(*)
from orderhdr 
where netpay > 1500 
group by year,month
order by year,month
```

## 2 สร้างชุดตำสั่งสืบค้นข้อมูลเพื่อแสดงวันที่ขายดีที่สุด (OrderHdr.Total) ในปี 2003

```python
%%sql
    select to_char(netpay,'9,999'),
    intime
    from orderhdr 
    where extract(year from intime)=2003 and netpay=(select max(netpay) from orderhdr where extract(year from intime)=2003)
    group by intime,netpay
```

## 3 สร้างคำสั่งสืบค้นข้อมูลเพื่อยอดขายสุทธิ (NetPay) ของวันหยุดสุดสัปดาห์ (เสาร์และอาทิตย์) ของปี 2003

```python
%%sql
select netpay,
    extract(dow from intime) "dow",
    to_char(intime,'Dy'),
    intime
from orderhdr
where extract(year from intime)=2003 and (extract(dow from intime)=0 or extract(dow from intime)=6)
order by intime
```

## 4 สร้างคำสั่งสืบค้นข้อมูลเพื่อหายอดขายแยกตามไตรมาสของปี 2003

```python
%%sql
with cte as(select ntile(4) over(),
    netpay
    from orderhdr)
select ntile,sum(netpay) 
from cte 
group by ntile 
order by ntile
    
```

## 5 สร้างคำสั่งสืบค้นข้อมูลเพื่อแสดงอาหารยอดนิยม (สำหรับ to-go) แยกเป็นรายเดือนของปี 2003

```python
%%sql
with cte as(
select orderid,
    menuid,
    descr,
    descrth,
    categoryid,
    isrecommended 
from menu inner join ordermenudtl using(menuid)
where isrecommended = True
),
cte2 as(
select intime,
    orderid,
    dinetypeid 
from orderhdr 
where dinetypeid = 'TG' and extract(year from intime)=2003
)
select * 
from cte inner join cte2 using(orderid)
order by intime
```

```python
%%sql
with cte as(
    select extract(month from intime)::int "month",
        dinetypeid,
        menuid,
        sum(qty) total
    from orderhdr inner join ordermenudtl using(orderid)
    where extract(year from intime)=2003 and dinetypeid='TG'
    group by month,dinetypeid,menuid
),
cte2 as(
    select *,rank() over(partition by month order by total desc) from cte
)
select menuid,descrth,month,total,dinetypeid
from cte2 inner join menu using(menuid) 
where rank=1 
group by menuid,descrth,total,dinetypeid,month
order by month
```

## 6 สร้างคำสั่งสืบค้นข้อมูลเพื่อแสดงอาหารที่ได้รับความนิยมน้อยที่สุด 3 อันดับแรก

```python
%%sql
select menuid,
    descrth,
    to_char(sum(qty),'999,999') total_order 
    from menu inner join ordermenudtl using(menuid) 
    group by menuid,descrth
    order by total_order 
    fetch first 3 rows only
```

## 7 สร้างคำสั่งสืบค้นข้อมูลเพื่อแสดงอาหารที่สร้างผลกำไรให้น้อยที่สุด 3 อันดับแรก

```python
%%sql
with cte as(
    select menuid,
    to_char(sum(netpay),'999,999,999.99') revenue
from orderhdr inner join ordermenudtl using(orderid) 
group by menuid
order by revenue
fetch first 3 rows only)
select menuid,descrth,revenue from cte inner join menu using (menuid) 
```

## 8 สร้างคำสั่งสืบค้นข้อมูลเพื่อหาว่าลูกค้ารายใดที่เข้ามารับประทานอาหารในช่วง 18:00 - 19:00 บ่อยครั้งที่สุด

```python
%%sql 
with cte as(select orderid,
    customerid,
    intime,
    outtime,
    count(customerid) "count"
from orderhdr 
where  -- to_char(intime,'HH24:MN:SS') between '18:00:00' and '19:00:00' -- ไม่รวม 18 19
    intime::time between '18:00' and '19:00' -- รวม 18 19
group by orderid,customerid,intime,outtime),
cte2 as(
select *,count(*) over(partition by customerid) "times" from cte)
select distinct customerid,Fname ||' '|| Lname "Name",times from cte2 inner join customer using(customerid) order by times desc 
```

## 9 สร้างคำสั่งสืบค้นข้อมูลเพื่อหาว่าลูกค้าที่เข้ามารับประทานอาหารในวันเสาร์และอาทิตย์บ่อยครั้งที่สุด

```python
%%sql
with cte as(select orderid,
    customerid,
    extract(dow from intime),
    count(*)
from orderhdr where extract(dow from intime) in (0,6) 
group by orderid,customerid,extract(dow from intime)
order by customerid),
cte2 as(
select *,sum(count) over(partition by customerid) "times" from cte)
select distinct กcustomerid,times from cte2 order by times desc
```

## 8 สร้างคำสั่งสืบค้นข้อมูลเพื่อหาว่ายอดขายสุทธิของวันหยุดสุดสัปดาห์เป็นกี่เปอร์เซ็นต์ของยอดขายสุทธิรวมในปี 2003

```python
%%sql
select to_char(100*sum(netpay)/
(select sum(netpay) from orderhdr where extract(year from intime) = 2003),'99.9999 %')  "weekend/total"
from orderhdr where extract(year from intime) = 2003 and extract(dow from intime) in (0,6)
```

 ## 9 สร้างคำสั่งสืบค้นข้อมูลเพื่อเปรียบเทียบยอดขายสุทธิของลูกค้าที่ซื้อไปทานนอกร้านเเละทานในร้านในแต่ละเดือนของปี 2003

```python
%%sql
select extract(month from intime)::int "month",dinetypeid, to_char(sum(netpay),'999,999,999.99') 
from orderhdr 
where extract(year from intime)=2003 
group by month,dinetypeid
order by dinetypeid
```

## 10 สร้างคำสั่งสืบค้นข้อมูลเพื่อแสดงว่าลูกค้าคนใดได้รับส่วนลดจากทางร้านน้อยที่สุด

```python
%%sql
select customerid,
    to_char(sum(discount),'999,999.99 บาท') "sum_of_discount" 
from orderhdr 
group by customerid
order by sum_of_discount
```

## 12 สร้างคำสั่งสืบค้นข้อมูลเพื่อสรุปประเภทการชำระเงินโดยเทียบเป็นเปอร์เซ็นต์ของยอดรวม (แสดงจำนวนครั้ง จำนวนเงิน)

```python
%%sql
with cte as(select paymentid,
            count(*) over(partition by paymentid) "times",
            sum(netpay) over(partition by paymentid) "total"
            from orderhdr
            order by paymentid),
cte2 as(select distinct *,round(total/(select sum(netpay) from orderhdr)*100,2) "%"from cte),
cte3 as(select 'Nothing',sum(times),sum(total),round(sum(total/(select sum(netpay) from orderhdr)*100),2) from cte2)
select * from cte2
union
select * from cte3
order by times
```

## 13 สร้างคำสั่งสืบค้นข้อมูลเพื่อแสดงเมนูที่มียอดขาย (จำนวนครั้งของการถูกสั่ง) ในเดือนพฤษภาคม 2003 เพิ่มขึ้นจากเดือนเดียวกันของปี 2002

```python
%%sql
with cte as(select menuid,sum(qty) "a",extract(month from intime) "aa"
from orderhdr inner join ordermenudtl using(orderid)
where extract(year from intime) = 2002
group by menuid,extract(month from intime)
order by menuid),
cte2 as(select  menuid,sum(qty) "b",extract(month from intime) "bb"
from orderhdr inner join ordermenudtl using(orderid)
where extract(year from intime) = 2003
group by menuid,extract(month from intime)
order by menuid)
select menuid,a,b from cte inner join cte2 using (menuid)
where aa =5 and bb=5 and b-a>0
```

## จงสร้างคำสั่งสืบค้นเพื่อหาเมนูที่ขายได้สูงสุด 3 อันดับแรกของแต่ละเดือน

```python
%%sql
with cte as(select extract(month from intime)::int "month",
            menuid,
            descrth,
            to_char(sum(qty),'999,999') "total_qty"
            from orderhdr inner join ordermenudtl using(orderid) inner join menu using(menuid)
            group by month,menuid,descrth
            order by month),
cte2 as (select *,rank() over(partition by month order by total_qty desc) from cte)
select * from cte2 where rank<4
```
