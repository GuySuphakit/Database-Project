## functions

````python
def list_table(con):
    '''
    list tables in connected database
    '''
    sql=f'''
    SELECT table_catalog, table_schema, table_name
        FROM information_schema.tables 
        WHERE table_type = 'BASE TABLE' 
            AND table_schema not in ('information_schema', 'pg_catalog')
        ORDER BY table_type, table_name
    '''
    return pd.read_sql(sql, con)

def describe_table(table_name, con):
    '''
    describe a table
    '''
    sql=f'''
    SELECT table_name, 
        case 
           when character_maximum_length is null then column_name
           else column_name || '(' || character_maximum_length || ')'
        end as col_name,    
        data_type, is_nullable, column_default default_value
        -- character_maximum_length, numeric_precision, datetime_precision,
    FROM 
       information_schema.columns
    WHERE 
       table_name = '{table_name}'
    ORDER BY table_name;
    '''
#     print(sql)
    return pd.read_sql(sql, con)

def list_db(con):
    '''
    list databases
    '''
    sql=f'''
    SELECT datname, datctype, datacl FROM pg_database order by datname;
    '''
#     print(sql)
    return pd.read_sql(sql, con)
````

````python
%%sql
select c.descr "category",count(*) "# menu"
    from menu m right join category c using(categoryid)
    group by "category"
    order by "# menu" desc
````

````python
%%sql
with t1 as(
    select c.descr "category",count(*) "# menu"
    from menu m inner join category c using(categoryid)
    group by "category"
    order by "# menu" desc
)
select c.descr,"# menu"
    from category c left join t1 on c.descr=t1.category
    where "# menu" is null
````

````python
%%sql
with t1 as(
select distinct menuid
    from orderhdr a inner join ordermenudtl b using(orderid)
    where intime::date = '2002-09-09'
)
select m.menuid "menuid", t1.menuid "t1 menuid", m.descrth 
    from menu m left join t1 using(menuid)
    where t1.menuid is null
````

````python
%%sql
select extract(month from intime)::int as "month",
        sum(Netpay) revenue,
        b.Area,
        (sum(Netpay)/b.Area)::decimal(8,2) "revenue per sq.m"
    from OrderHdr a cross join Parameter b
    where extract(year from intime)= 2002
    group by month, b.Area
    order by month
````

````python
%%sql
with y02 as(
    select extract(month from intime)::int "month",sum(netpay) revenue
    from orderhdr
    where extract(year from intime)=2002
    group by "month"
),
y03 as(
    select extract(month from intime)::int "month",sum(netpay) revenue
    from orderhdr
    where extract(year from intime)=2003
    group by "month"
)
select y02.month, y02.revenue "2002", y03.revenue "2003",
    y03.revenue-y02.revenue "2003-2002"
    from y02 inner join y03 using(month)
````

````python
%%sql
select CustomerID, Fname||''||Lname "Full name", Gender,
        case
            when Gender ='F' then 'Ms.' --''ขีดเดียวจะพูดถึงตัวแปล 
            when Gender ='M' then 'Mr.' --flexible กว่า ใช้ and เพิ่มเงื่อนไขได้
        end  || '' ||Fname||''||Lname as title
    from Customer
````

````python
%%sql
-- Q:แสดงคำนำหน้าชื่อ
-- Syntax 1: [column name] = case ... end
select CustomerID, Fname||''||Lname "Full name", Gender,
        case Gender
            when 'F' then 'Ms.' 
            when 'M' then 'Mr.'
        end  || '' ||Fname||''||Lname as title
    from Customer
````
